```python
df = pd.read_excel('Contribuição SMILES em Tasks_COM FIGURA.xlsx')
df
```

```python
df['Heavy Atom Count'] = [HeavyAtomCount(Chem.MolFromSmiles(smiles)) for smiles in df['Fragment SMILES']]
df.head()
```

```python
PandasTools.AddMoleculeColumnToFrame(df,'Fragment SMILES','Molecule')
df.head()
```

```python
df = df.sort_values(by=['Heavy Atom Count'],ascending=False)
df.head()
```

```python
df = df.drop(columns = 'Unnamed: 0')
df.head()
```

```python
df.shape
```

```python
df = df.loc[(df['Heavy Atom Count']) >= 7]
df
```

```python
df.shape
```

```python
df = df.sort_values(by=['Task'],ascending=True)
df
```

```python
df['smarts'] = [Chem.MolToSmarts(Chem.MolFromSmiles(smiles)) for smiles in df['Fragment SMILES']]
df.head()
```

```python
df0 = df.loc[(df['Task']) == 0]
df0
```

```python
df['Canonical'] = [Chem.CanonSmiles(smiles) for smiles in df['Fragment SMILES']]
df.head()
```

```python
df.shape
```

```python
result_df = df.drop_duplicates(subset=['Canonical'], keep='first')
result_df.shape
```

```python
result_df
```

```python
result_df0 = result_df.loc[(result_df['Task']) == 0]
result_df0
```

```python
toxic_alert_list = result_df['Fragment SMILES'].tolist()

mol_list = []
for smiles in toxic_alert_list:
  mol = Chem.MolFromSmiles(smiles)
  mol_list.append(mol)
img =Draw.MolsToGridImage(mol_list, molsPerRow = 10, maxMols = 150)
img  
```

```python
dftop5Tasks = pd.read_excel('Highlights Top 5_COM FIGURA.xlsx')
dftop5Tasks
```

```python
dftop5Tasks['Heavy Atom Count'] = [HeavyAtomCount(Chem.MolFromSmiles(smiles)) for smiles in dftop5Tasks['Fragment SMILES']]
dftop5Tasks.head()
```

```python
dftop5Tasks = dftop5Tasks.loc[(dftop5Tasks['Heavy Atom Count']) >= 7]
dftop5Tasks.reset_index(inplace=True)
dftop5Tasks
```

```python
df_molecules = pd.read_excel('test_dataset_molecules.xlsx')
df_molecules
```

```python
lHighlights = []
for (i,lmolecules) in enumerate(dftop5Tasks['Active Molecules'].values):
    index_mol = int(lmolecules.split(',')[0][1:].replace(']',''))
    molecule_smile = df_molecules['ids'][index_mol]
    m = Chem.MolFromSmiles(molecule_smile)
    frag = dftop5Tasks['Fragment SMILES'][i]
    substructure = Chem.MolFromSmiles(frag)
    m.GetSubstructMatches(substructure)
    lHighlights.append(m)

df_highlight = pd.DataFrame(lHighlights)
df_highlight
```
